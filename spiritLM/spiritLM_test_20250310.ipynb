```python
from spiritlm.model.spiritlm_model import Spiritlm, OutputModality, GenerationInput, ContentType
from transformers import GenerationConfig
```

```python
spirit_lm = Spiritlm("spirit-lm-expressive-7b")
```

## Text test

```python
import time
import numpy as np


test_ans = []
test_time = []
def text_test():
    time_start = time.time()
    output = spirit_lm.generate(
        output_modality=OutputModality.TEXT,
        interleaved_inputs=[
            GenerationInput(
                content="The largest country in the world is",
                content_type=ContentType.TEXT,
            )
            ],
        generation_config=GenerationConfig(
            temperature=0.9,
            top_p=0.95,
            max_new_tokens=50,
            do_sample=True,
        ),
    )
    
    time_end = time.time()
    spending_time = time_end - time_start
    
    return spending_time, output[0].content

for i in range(10):
    t, o = text_test()
    test_ans.append(o)
    test_time.append(t)
    
print(test_ans)
print(np.average(np.array(test_time)))
```

## Text-to-Speech test

```python
import IPython.display as ipd

def display_outputs(outputs):
    for output in outputs:
        if output.content_type == ContentType.TEXT:
            print(output.content)
        else:
            ipd.display(ipd.Audio(output.content, rate=16_000))
            
```

```python
import time
time_start = time.time()
outputs2 = spirit_lm.generate(
    output_modality=OutputModality.SPEECH,
    interleaved_inputs=[('text', "I am so deeply saddened, it feels as if my heart is shattering into a million pieces and I can't hold back the tears that are streaming down my face.")],
    generation_config=GenerationConfig(
        temperature=0.9,
        top_p=0.95,
        max_new_tokens=200,
        do_sample=True,
    ),
    speaker_id=1,
)

time_end = time.time()

display_outputs(outputs2)
print(time_end-time_start)
```

```python
import soundfile as sf

sf.write("output_text2speech.wav", outputs2[0].content, 16_000)
```
